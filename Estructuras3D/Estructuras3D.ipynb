### Ejercicio 1. 
Descarga la estructura de la proteína de la hemoglobina humana (código PDB: 1A3N), visualízala, y carga el archivo en Biopython (creando un objeto estructura).

(a) Calcula la distancia entre los átomos CA O del primer y último residuo de la cadena A de la hemoglobina.

(b) Calcula el ángulo diedro entre los átomos N,CA,C, y O del primer residuo de la cadena A.

(c) Calcula el centro de masas de la estructura de la hemoglobina.

```python
from Bio.PDB.MMCIFParser import MMCIFParser
from Bio.PDB import Vector
import numpy as np

parser = MMCIFParser()
structure = parser.get_structure('hemoglobina', '1a3n.cif')

last_residue = None
for model in structure:
    for chain in model:
        if chain.id == 'A':
            last_residue = list(chain)[-1]  

if last_residue:
    atom_CA_first = structure[0]['A'][1]['CA']
    atom_O_last = last_residue['O']

    distancia = atom_CA_first - atom_O_last
    print(f"Distancia entre CA del primer y O del último residuo de la cadena A: {distancia:.2f} Å")
else:
    print("No se pudo encontrar el último residuo en la cadena A.")

###########################################

atom_N = structure[0]['A'][1]['N']
atom_CA = structure[0]['A'][1]['CA']
atom_C = structure[0]['A'][1]['C']
atom_O = structure[0]['A'][1]['O']

coord_N = atom_N.get_vector().get_array()
coord_CA = atom_CA.get_vector().get_array()
coord_C = atom_C.get_vector().get_array()
coord_O = atom_O.get_vector().get_array()

#dihedral
v1 = coord_N - coord_CA
v2 = coord_C - coord_CA
v3 = coord_C - coord_O

angle = np.arccos(np.dot(np.cross(v1, v2), np.cross(v2, v3)) / (np.linalg.norm(np.cross(v1, v2)) * np.linalg.norm(np.cross(v2, v3))))
print(f"Ángulo diedro entre N, CA, C, O del primer residuo de la cadena A: {angle:.2f} radianes")

#############################################

center_of_mass = Vector(0, 0, 0)
total_atoms = 0

for model in structure:
    for chain in model:
        for residue in chain:
            for atom in residue:
                center_of_mass += atom.get_vector()
                total_atoms += 1

center_of_mass /= total_atoms
print(f"Centro de masas de la estructura de la hemoglobina: {center_of_mass.get_array()} Å")
```

### Ejercicio 2. 
Descarga la estructura de la proteína lisozima del huevo de gallina (código PDB: 1LYZ), visualízala, y carga el archivo en Biopython (creando un objeto estructura).

(a) Imprima el número de átomos y proporcione el nombre del primer y último átomo de la lista.

(b) Calcula el ángulo entre los tres primeros átomos de la lista y muestre su valor en grados.

(c) Muestre por pantalla la cadena y el residuo del átomo que se encuentra justo en la mitad de la lista.

```python
from Bio.PDB.MMCIFParser import MMCIFParser
import numpy as np

parser = MMCIFParser()
structure = parser.get_structure('lisozima', '1lyz.cif')  # Reemplaza '1LYZ.pdb' con tu archivo descargado

atom_list = list(structure.get_atoms())
num_atoms = len(atom_list)

first_atom_name = atom_list[0].get_name()
last_atom_name = atom_list[-1].get_name()

print(f"Número de átomos en la estructura: {num_atoms}")
print(f"Nombre del primer átomo: {first_atom_name}")
print(f"Nombre del último átomo: {last_atom_name}")

if num_atoms >= 3:
    atom1 = atom_list[0].get_vector()
    atom2 = atom_list[1].get_vector()
    atom3 = atom_list[2].get_vector()

    v1 = atom1 - atom2
    v2 = atom3 - atom2

    angle_rad = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))
    angle_deg = np.degrees(angle_rad)
    print(f"Ángulo entre los tres primeros átomos: {angle_deg:.2f} grados")
else:
    print("La estructura no tiene suficientes átomos para calcular el ángulo.")

mid_atom_index = num_atoms // 2
if mid_atom_index < num_atoms:
    mid_atom = atom_list[mid_atom_index]
    residue = mid_atom.get_parent()
    chain_id = residue.get_parent().id

    print(f"Átomo en la mitad de la lista:")
    print(f"    Cadena: {chain_id}")
    print(f"    Residuo: {residue}")
else:
    print("No se puede encontrar un átomo en la mitad de la lista.")
```

(e) [OPCIONAL] Desarrolle una función con Biopython que dada una estructura devuelva un vector, que proporcione, para cada modelo, la distancia promedio entre sus átomos.       Utilícela para la Hipocretina-1/Orexina-A y la Orexina-B/Hipocretina-2 e imprima los vectores resultados por pantalla.

```python
from Bio.PDB import MMCIFParser
from Bio.SeqUtils import seq1

parser = MMCIFParser()
structure_1WSO = parser.get_structure("1WSO", "1wso.cif")
structure_1CQ0 = parser.get_structure("1CQ0", "1cq0.cif")

sequence_1WSO = ''.join(seq1(residue.get_resname()) for model in structure_1WSO for chain in model for residue in chain if residue.get_id()[0] == ' ')
sequence_1CQ0 = ''.join(seq1(residue.get_resname()) for model in structure_1CQ0 for chain in model for residue in chain if residue.get_id()[0] == ' ')

min_length = min(len(sequence_1WSO), len(sequence_1CQ0))
sequence_1WSO = sequence_1WSO[:min_length]
sequence_1CQ0 = sequence_1CQ0[:min_length]

with open("1WSO.fasta", "w") as file_1WSO:
    file_1WSO.write(f">1WSO\n{sequence_1WSO}\n")

with open("1CQ0.fasta", "w") as file_1CQ0:
    file_1CQ0.write(f">1CQ0\n{sequence_1CQ0}\n")

from Bio import pairwise2

alignments = pairwise2.align.globalxx(sequence_1WSO, sequence_1CQ0)
alignment = alignments[0]
similarity = (alignment.score / max(len(sequence_1WSO), len(sequence_1CQ0))) * 100

print(f"Alineamiento:\n{alignment}\n")
print(f"Grado de similitud: {similarity:.2f}%")


```

```python
from Bio.Blast import NCBIWWW, NCBIXML

def blast_search(query_sequence):
    result_handle = NCBIWWW.qblast("blastp", "nr", query_sequence)
    blast_records = NCBIXML.parse(result_handle)
    return blast_records

blast_records_1WSO = blast_search(sequence_1WSO)
blast_records_1CQ0 = blast_search(sequence_1CQ0)

def show_blast_results(blast_records):
    for blast_record in blast_records:
        for alignment in blast_record.alignments:
            print(f"Secuencia: {alignment.title}")

print("Resultados BLAST para 1WSO:")
show_blast_results(blast_records_1WSO)

print("\nResultados BLAST para 1CQ0:")
show_blast_results(blast_records_1CQ0)
```

```python
from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor
from Bio import AlignIO, Phylo

with open("alignments.fasta", "w") as alignments_file:
    alignments_file.write(f">1WSO\n{sequence_1WSO}\n")
    alignments_file.write(f">1CQ0\n{sequence_1CQ0}\n")

alignment = AlignIO.read("alignments.fasta", "fasta")

calculator = DistanceCalculator('blosum62')
dm = calculator.get_distance(alignment)
constructor = DistanceTreeConstructor()
tree = constructor.nj(dm)

Phylo.draw(tree)
```

```python


def average_model_distances(structure):
    average_distances = []
    for model in structure:
        atoms = list(model.get_atoms())
        num_atoms = len(atoms)
        total_distance = 0
        
        # Calculate pairwise distances between atoms
        for i, atom1 in enumerate(atoms):
            for atom2 in atoms[i+1:]:
                total_distance += atom1 - atom2  # Calculate distance using Bio.PDB

        # Calculate average distance for the model
        average_distance = total_distance / (num_atoms * (num_atoms - 1) / 2)
        average_distances.append(average_distance)
    
    return average_distances

avg_distances_h1_oa = average_model_distances(structure_1WSO)
avg_distances_ob_h2 = average_model_distances(structure_1CQ0)

print("Average distances for Hipocretina-1/Orexina-A:")
for i, distance in enumerate(avg_distances_h1_oa, start=1):
    print(f"Model {i}: {distance:.4f} Å")

print("\nAverage distances for Orexina-B/Hipocretina-2:")
for i, distance in enumerate(avg_distances_ob_h2, start=1):
    print(f"Model {i}: {distance:.4f} Å")
```
