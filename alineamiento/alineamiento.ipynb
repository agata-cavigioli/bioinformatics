# ejercicio 1

```python
from Bio import Align
from Bio.Seq import Seq
from Bio import SeqIO
import random

print('SECUENCIAS GENERADA')
bio_seq1 = Seq(''.join(random.choice('ACGT') for _ in range(1,50)))
bio_seq2 = Seq(''.join(random.choice('ACGT') for _ in range(1,50)))

print("Secuencia generada 1:", bio_seq1)
print("Secuencia generada 2:", bio_seq2)



aligner = Align.PairwiseAligner()

alignments = aligner.align(bio_seq1, bio_seq2)
best_alignment = alignments[0]
print("\n ALINEAMIENTO OPTIMO:")
print(best_alignment)
print("score:", best_alignment.score)

#hemoglobin
#insuline
registros = list(SeqIO.parse("dna.fasta", "fasta"))
print('\n SECUENCIAS DEL FICHERO')
count = 0
for record in registros:
    count += 1
    
    print("Secuencia del fichero %i:" % count,record.seq)

bio_seq3 = registros[0].seq
bio_seq4 = registros[1].seq

aligner = Align.PairwiseAligner()

alignments_2 = aligner.align(bio_seq3, bio_seq4)

alignments_3 = alignments_2[0]
print("\n ALINEAMIENTO OPTIMO:")
print(alignments_3)
print("score:", alignments_3.score)


aligner.mode = 'local'  # 'global' or 'local' alignment
aligner.match_score = 2  # Score for match
aligner.mismatch_score = -1  # Score for mismatch
aligner.open_gap_score = -0.5  # Score for opening a gap
aligner.extend_gap_score = -0.1  # Score for extending a gap


print("\n PARAMETROS MODIFICADOS:")
print(aligner)
print()

a = aligner.align(bio_seq1, bio_seq2)
a_1 = a[0]
print("score de las secuencias generada:", a_1.score)

a_2 = aligner.align(bio_seq3, bio_seq4)
a_3 = a_2[0]
print("score de las secuencias del fichero:", a_3.score)
```

### Comentario

El código utiliza BioPython para realizar alineamientos de secuencias de ADN. Comienza generando dos secuencias aleatorias de longitud 50, luego las alinea y muestra el mejor alineamiento junto con su puntuación. Posteriormente, carga secuencias desde un archivo FASTA ("dna.fasta"), las alinea entre sí y muestra el mejor alineamiento con su puntuación.

Después, modifica los parámetros del alineador para realizar alineamientos locales, ajustando puntajes para coincidencias, no coincidencias y brechas, y muestra los parámetros modificados. Realiza alineamientos nuevamente para las secuencias generadas aleatoriamente y las del archivo FASTA, mostrando sus respectivas puntuaciones.

El comentario sobre la salida obtenida es que los alineamientos muestran las regiones donde las secuencias coinciden y difieren, con líneas verticales que indican coincidencias y líneas horizontales que representan brechas. Las puntuaciones reflejan la similitud entre las secuencias, donde puntuaciones más altas indican una mayor similitud entre las secuencias alineadas. En este caso, las secuencias del archivo FASTA muestran una similitud considerablemente mayor en comparación con las secuencias generadas aleatoriamente.

# ejercicio 2

```python
from random import randint, choice

print('SECUENCIAS GENERADA')
seq1 = Seq("".join([choice("ACDEFGHIKLMNPQRSTVWY") for _ in range(randint(5, 75))]))
seq2 = Seq("".join([choice("ACDEFGHIKLMNPQRSTVWY") for _ in range(randint(5, 75))]))

# Imprimir las secuencias generadas
print("Secuencia 1:", seq1)
print("Secuencia 2:", seq2)

aligner = Align.PairwiseAligner()

alignments = aligner.align(seq1, seq2)

alignments_1 = alignments[0]
print("\n ALINEAMIENTO OPTIMO:")
print(alignments_1)
print("score:", alignments_1.score)
print()
print()

#myelin basic protein
#Myelin protein P0
file_path = 'protein.fasta'  # Reemplaza con tu ruta

# Leer el archivo FASTA y obtener las secuencias
sequences = []
for seq_record in SeqIO.parse(file_path, 'fasta'):
    sequences.append(seq_record.seq)
seq3 = sequences[0]
seq4 = sequences[1]

print('\n SECUENCIAS DEL FICHERO')

print("Secuencia 1:", seq3)
print("Secuencia 2:", seq4)

aligner = Align.PairwiseAligner()

alignments_2 = aligner.align(seq3, seq4)

alignments_3 = alignments_2[0]
print("\n ALINEAMIENTO OPTIMO:")
print(alignments_3)
print("score:", alignments_3.score)
print()
print()

from Bio.Align import substitution_matrices
aligner_2 = Align.PairwiseAligner()
aligner_2.substitution_matrix = substitution_matrices.load("BLOSUM62")

print('\n BLOSUM')

a = aligner_2.align(seq1, seq2)
a_1 = a[0]
print("score de las secuencias generada:", a_1.score)

a_2 = aligner_2.align(seq3, seq4)
a_3 = a_2[0]
print("score de las secuencias del fichero:", a_3.score)
```

### Comentario
El código de nuevo genera dos secuencias aleatorias y luego realiza un alineamiento entre ellas para mostrar el mejor alineamiento y su puntuación. Después, carga secuencias desde un archivo FASTA ("protein.fasta") y realiza un alineamiento entre ellas, mostrando el mejor alineamiento y su puntuación.

Posteriormente, utiliza una matriz de sustitución específica (en este caso, la matriz BLOSUM62) para realizar alineamientos entre las secuencias generadas aleatoriamente y las del archivo FASTA, mostrando las puntuaciones resultantes. La matriz BLOSUM62 es comúnmente utilizada en bioinformática para alinear secuencias de proteínas y evaluar la similitud entre ellas. Sus puntajes se basan en la frecuencia observada de sustituciones de aminoácidos en secuencias evolutivamente relacionadas.

La puntuación refleja la similitud entre las secuencias alineadas, donde puntuaciones más altas indican mayor similitud. Además, se incluye la comparación de puntuaciones utilizando la matriz BLOSUM62 para resaltar cómo cambian las puntuaciones al utilizar un enfoque diferente para el alineamiento.
