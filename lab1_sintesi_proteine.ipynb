```python
import random
import numpy as np

sequence_length = 50

def generate_random_dna_sequence(length):
    dna_bases = ['A', 'T', 'C', 'G']
    start_sequence = 'TAC'  # Start codon
    stop_sequences = ['ATT', 'ATC', 'ACT']
    random_stop_sequence = random.choice(stop_sequences)
    
    random_sequence = ''.join(random.choice(dna_bases) for _ in range(length-6))
    seq =  start_sequence + random_sequence + random_stop_sequence
    
    DNA_sequence_complete = [['' for _ in range(2)] for _ in range(sequence_length)]
    conversion_dict1 = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    i = 0
    for base in seq:
        opp_base = conversion_dict1[base]
        DNA_sequence_complete[i][0] = base
        DNA_sequence_complete[i][1] = opp_base
        i+=1
    return DNA_sequence_complete


DNA_sequence_complete = generate_random_dna_sequence(sequence_length)
print("DNA Sequence Complete:", DNA_sequence_complete)


column = [row[0] for row in DNA_sequence_complete]
#print(column)
DNA_sequence = ''.join(column)

#print("DNA Sequence:", DNA_sequence)


'''
Replication:

Purpose: Replication is the process by which a cell makes an exact copy of its DNA. 
It occurs during the cell division process, such as mitosis or meiosis, to ensure that 
each daughter cell receives a complete set of genetic instructions.
Key Steps:
DNA unwinding: The double-stranded DNA molecule unwinds, exposing its two complementary strands.
Complementary base pairing: Free nucleotides in the cell's nucleus align with their complementary 
bases on the exposed DNA strands. Adenine (A) pairs with thymine (T), and cytosine (C) pairs with guanine (G).
DNA polymerase: Enzymes, such as DNA polymerase, help in assembling the new strands by joining 
the complementary nucleotides. Two identical DNA molecules are formed, each with one original 
and one newly synthesized strand.

'''
print('\nREPLICATION')

def replication(DNA_sequence_complete):
    column = [row[0] for row in DNA_sequence_complete]
    DNA_sequence = ''.join(column)

    print("DNA half sequence:", DNA_sequence)
    
    DNA_sequence_copy = ['' for _ in range(sequence_length)] 
    conversion_dict1 = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    i = 0
    for base in DNA_sequence:
        opp_base = conversion_dict1[base]
        DNA_sequence_copy[i] = opp_base
        i+=1
        
    DNA_sequence_complete_copy = [['' for _ in range(2)] for _ in range(sequence_length)]
    i = 0
    for base in DNA_sequence_copy:
        opp_base = conversion_dict1[base]
        DNA_sequence_complete_copy[i][1] = base
        DNA_sequence_complete_copy[i][0] = opp_base
        i+=1
    return DNA_sequence_complete_copy

DNA_copy = replication(DNA_sequence_complete)
print("DNA Sequence Complete:", DNA_copy)
'''
Transcription:

Purpose: Transcription is the process by which a specific segment of DNA is used as a template to 
synthesize a complementary RNA molecule. This RNA molecule, called messenger RNA (mRNA), carries 
the genetic information from DNA to the ribosomes for protein synthesis.
Key Steps:
Initiation: RNA polymerase, an enzyme, binds to a specific promoter region on the DNA molecule.
Elongation: RNA polymerase moves along the DNA template strand and synthesizes a complementary mRNA 
strand by adding nucleotides.
Termination: Transcription concludes when the RNA polymerase reaches a termination signal on the DNA.

'''
print('\nTRANSCRIPTION')
# Define the mapping of codons to amino acids (genetic code)
genetic_code = {
    'UUU': 'Phenylalanine',
    'UUC': 'Phenylalanine',
    'UUA': 'Leucine',
    'UUG': 'Leucine',
    'CUU': 'Leucine',
    'CUC': 'Leucine',
    'CUA': 'Leucine',
    'CUG': 'Leucine',
    'AUU': 'Isoleucine',
    'AUC': 'Isoleucine',
    'AUA': 'Isoleucine',
    'AUG': 'Methionine',  # Start codon
    'GUU': 'Valine',
    'GUC': 'Valine',
    'GUA': 'Valine',
    'GUG': 'Valine',
    'UCU': 'Serine',
    'UCC': 'Serine',
    'UCA': 'Serine',
    'UCG': 'Serine',
    'CCU': 'Proline',
    'CCC': 'Proline',
    'CCA': 'Proline',
    'CCG': 'Proline',
    'ACU': 'Threonine',
    'ACC': 'Threonine',
    'ACA': 'Threonine',
    'ACG': 'Threonine',
    'GCU': 'Alanine',
    'GCC': 'Alanine',
    'GCA': 'Alanine',
    'GCG': 'Alanine',
    'UAU': 'Tyrosine',
    'UAC': 'Tyrosine',
    'UAA': 'Stop',
    'UAG': 'Stop',
    'CAU': 'Histidine',
    'CAC': 'Histidine',
    'CAA': 'Glutamine',
    'CAG': 'Glutamine',
    'AAU': 'Asparagine',
    'AAC': 'Asparagine',
    'AAA': 'Lysine',
    'AAG': 'Lysine',
    'GAU': 'Aspartic Acid',
    'GAC': 'Aspartic Acid',
    'GAA': 'Glutamic Acid',
    'GAG': 'Glutamic Acid',
    'UGU': 'Cysteine',
    'UGC': 'Cysteine',
    'UGA': 'Stop',
    'UGG': 'Tryptophan',
    'CGU': 'Arginine',
    'CGC': 'Arginine',
    'CGA': 'Arginine',
    'CGG': 'Arginine',
    'AGU': 'Serine',
    'AGC': 'Serine',
    'AGA': 'Arginine',
    'AGG': 'Arginine',
    'GGU': 'Glycine',
    'GGC': 'Glycine',
    'GGA': 'Glycine',
    'GGG': 'Glycine',
}


def transcription(dna_sequence):
    conversion_dict = {'A': 'U', 'T': 'A', 'C': 'G', 'G': 'C'}
    mrna_sequence = ''
    current_codon = ''
    # Iterate through the DNA sequence and perform the conversion
    for base in dna_sequence:
        # Use the conversion dictionary to replace bases
        if base in conversion_dict:
            mrna_sequence += conversion_dict[base]
            #stops if reaches stop codon
            current_codon += conversion_dict[base]
            if len(current_codon) == 3:
                amino_acid = genetic_code.get(current_codon, 'Unknown')
                if amino_acid == 'Stop':
                    break  # Stop translation at a stop codon
                current_codon = ''
            ####
        else:
            # If the base is not A, T, C, or G, it is considered an error
            print(f"Error: Invalid base '{base}' in DNA sequence.")
            return None
    
    return mrna_sequence


mRNA_sequence = transcription(DNA_sequence)

if mRNA_sequence is not None:
    print("mRNA Sequence:", mRNA_sequence)


'''
Translation:

Purpose: Translation is the process by which the information encoded in the mRNA is used to build a protein. 
It occurs in the ribosomes, cellular structures that serve as protein factories.
Key Steps:
Initiation: The ribosome assembles on the mRNA at the start codon (AUG), and the first transfer RNA (tRNA) 
carrying the amino acid methionine binds to the start codon.
Elongation: The ribosome moves along the mRNA in a 5' to 3' direction. Each tRNA, bearing its specific amino acid,
binds to the mRNA codon via complementary base pairing.
Peptide bond formation: As the ribosome moves, it catalyzes the formation of peptide bonds between adjacent amino 
acids, creating a growing polypeptide chain.
Termination: Translation concludes when the ribosome encounters a stop codon (UAA, UAG, or UGA). At this point,
a release factor binds to the ribosome, and the newly synthesized protein is released.
In summary, replication ensures the faithful copying of DNA, transcription converts DNA into mRNA, and 
translation uses the information in mRNA to synthesize proteins. These processes are vital for the flow of genetic information and the production of functional proteins in living cells.

'''
print('\nTRANSLATION')

def translation(mRNA_sequence):
    protein_sequence = []
    current_codon = ''
    for base in mRNA_sequence:
        current_codon += base
        if len(current_codon) == 3:
            amino_acid = genetic_code.get(current_codon, 'Unknown')
            if amino_acid == 'Stop':
                break  # Stop translation at a stop codon
            protein_sequence.append(amino_acid)
            current_codon = ''
    return protein_sequence

protein_sequence = translation(mRNA_sequence)
print("Protein Sequence:", ', '.join(protein_sequence))
```
